```python
from substack_api import Category
from substack_api.category import list_all_categories
```

# List categories and dig into specific one

```python
categories = list_all_categories()
print(categories)
```

```python
# get finance category
cat = Category(name="Finance")
```

```python
print(cat.name)
print(cat.id)
newsletters = cat.get_newsletters()
```
